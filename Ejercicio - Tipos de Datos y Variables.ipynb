```python
my_option_1 = 0 # Limpiando mi variable de selección para correr mi calculadora
while my_option_1 != 5:
    print("""
        Selecciona una opción para el tipo de numeración que usarás:

        1) Decimal (10)
        2) Binaria (0b, 0B)
        3) Octal (0o, 0O)
        4) Hexadecimal (0x,0X)
        5) Salir
        """)
    my_option_1 = int(input("Elige una opción: ") )

    if my_option_1 == 1:
        my_num1 = float(input("Introduce tu primer número (Decimal): "))
        my_num2 = float(input("Introduce tu segundo número (Decimal): "))
    elif my_option_1 == 2:
        my_num1 = int(input("Introduce tu primer número (Binario): "),2)
        my_num2 = int(input("Introduce tu segundo número (Binario): "),2)
    elif my_option_1 == 3:
        my_num1 = int(input("Introduce tu primer número (Octal): "),8)
        my_num2 = int(input("Introduce tu segundo número (Octal): "),8)
    elif my_option_1 == 4:
        my_num1 = int(input("Introduce tu primer número (Hexadecimal): "),16)
        my_num2 = int(input("Introduce tu segundo número (Hexadecimal): "),16)
    
    if my_option_1 > 5:
        print("Esa opción no está disponible")
    
    if my_option_1 < 5:

        my_option_2 = 0 # Limpiando mi variable de selección para correr mi calculadora

        print("""
            Selecciona una opción

            1) Sumar los dos números
            2) Restar los dos números
            3) Multiplicar los dos números
            4) Dividir los dos números
            5) Elevar a la potencia el primer número por el segundo número
            6) Salir
            """)
        my_option_2 = int(input("Elige una opción: ") )     
        if my_option_2 == 1:
            print("\n La suma de",my_num1,"+",my_num2,"es igual a",my_num1+my_num2)
            if my_option_1 == 1:
                print ("\n en decimal: ",my_num1+my_num2)
            elif my_option_1 == 2:
                print ("\n en binario: ",bin(my_num1+my_num2))
            elif my_option_1 == 3:
                print ("\n en octal: ",oct(my_num1+my_num2))
            elif my_option_1 == 4:
                print ("\n en hexadecimal: ",hex(my_num1+my_num2))
        elif my_option_2 == 2:
            print("\n La resta de",my_num1,"-",my_num2,"es igual a",my_num1-my_num2)
            if my_option_1 == 1:
                print ("\n en decimal: ",my_num1-my_num2)
            elif my_option_1 == 2:
                print ("\n en binario: ",bin(my_num1-my_num2))
            elif my_option_1 == 3:
                print ("\n en octal: ",oct(my_num1-my_num2))
            elif my_option_1 == 4:
                print ("\n en hexadecimal: ",hex(my_num1-my_num2))
        elif my_option_2 == 3:
            print("\n La multiplicación de",my_num1,"*",my_num2,"es igual a",my_num1*my_num2)
            if my_option_1 == 1:
                print ("\n en decimal: ",my_num1*my_num2)
            elif my_option_1 == 2:
                print ("\n en binario: ",bin(my_num1*my_num2))
            elif my_option_1 == 3:
                print ("\n en octal: ",oct(my_num1*my_num2))
            elif my_option_1 == 4:
                print ("\n en hexadecimal: ",hex(my_num1*my_num2))
        elif my_option_2 == 4:
            print("\n La división de",my_num1,"/",my_num2,"es igual a",my_num1/my_num2)
        elif my_option_2 == 5:
            print("\n La potencia de",my_num1,"**",my_num2,"es igual a",my_num1**my_num2)
            if my_option_1 == 1:
                print ("\n en decimal: ",my_num1**my_num2)
            elif my_option_1 == 2:
                print ("\n en binario: ",bin(my_num1**my_num2))
            elif my_option_1 == 3:
                print ("\n en octal: ",oct(my_num1**my_num2))
            elif my_option_1 == 4:
                print ("\n en hexadecimal: ",hex(my_num1**my_num2))
        elif my_option_2 == 6:
            print("Gracias por usar la calculadora")
        else:
            print("Opción incorrecta")
```
